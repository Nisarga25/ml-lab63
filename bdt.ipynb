```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay


iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=1)

clf_iris = GaussianNB()
clf_iris.fit(X_train, y_train)
y_pred_iris = clf_iris.predict(X_test)

accuracy_iris = accuracy_score(y_test, y_pred_iris)
print(f"Iris - Bayesian Classification Accuracy: {accuracy_iris*100:.2f}%")
cm_iris = confusion_matrix(y_test, y_pred_iris)
disp_iris = ConfusionMatrixDisplay(confusion_matrix=cm_iris, display_labels=iris.target_names)
disp_iris.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix (Iris Dataset)')
plt.show()

wine = datasets.load_wine()
X_train2, X_test2, y_train2, y_test2 = train_test_split(wine.data, wine.target, test_size=0.3, random_state=1)

clf_wine = GaussianNB()
clf_wine.fit(X_train2, y_train2)
y_pred_wine = clf_wine.predict(X_test2)

accuracy_wine = accuracy_score(y_test2, y_pred_wine)
print(f"Wine - Bayesian Classification Accuracy: {accuracy_wine*100:.2f}%")
cm_wine = confusion_matrix(y_test2, y_pred_wine)
disp_wine = ConfusionMatrixDisplay(confusion_matrix=cm_wine, display_labels=wine.target_names)
disp_wine.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix (Wine Dataset)')
plt.show()
```
